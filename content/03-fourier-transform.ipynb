# Fast Fourier Transform
***
The Fast Fourier Transform (FFT) is an algorithm that is used to quickly compute the discrete Fourier transform (DFT) of a sequence. The DFT is a mathematical operation that converts a sequence of complex numbers into another sequence of complex numbers, where the new sequence represents the frequency components of the original sequence. The FFT algorithm is used to speed up the computation of the DFT, and it has many applications in signal processing, image processing, and other areas of computational science.

The FFT algorithm works by decomposing a sequence of complex numbers into a series of simpler sequences, called "butterflies," which can be computed more quickly. Each butterfly consists of two complex numbers that are multiplied by a special set of coefficients, called "twiddle factors," which are chosen to simplify the computation. By decomposing the original sequence into a series of butterflies, the FFT algorithm is able to compute the DFT much more quickly than other methods.

One of the main advantages of the FFT algorithm is that it has a time complexity of O(n log n), which means that the computation time grows only logarithmically with the length of the input sequence. This makes the FFT algorithm much faster than other algorithms for computing the DFT, such as the naive algorithm, which has a time complexity of O(n^2).

There are many different variations of the FFT algorithm, including the Cooley-Tukey FFT, the Bluestein's FFT, and the Rader's FFT. Each of these variations has its own strengths and weaknesses, and they are used in different applications depending on the requirements of the problem. Overall, the FFT algorithm is a powerful and widely used tool for quickly computing the DFT of a sequence.

## Complex Numbers
***
Complex numbers are numbers that are made up of a real part and an imaginary part. The imaginary part is denoted by the symbol "i", which represents the square root of -1. The real part is a normal number, such as 1, 2, 3, etc. A complex number is written in the form a + bi, where a is the real part and bi is the imaginary part.

For example, the complex number 3 + 4i has a real part of 3 and an imaginary part of 4i. The real part is 3 and the imaginary part is 4 times the square root of -1.

Complex numbers can be added, subtracted, multiplied, and divided just like normal numbers. However, there are a few extra rules to keep in mind when working with complex numbers. For example, when multiplying complex numbers, you have to use the distributive property and the rule that i^2 = -1.

Complex numbers are used in many areas of mathematics, including calculus, linear algebra, and quantum mechanics. They are also used in electrical engineering and other fields where sinusoidal functions are important.

```python
# You can perform arithmetic operations on complex numbers just like you would with regular numbers. For example:
a = 2 + 3j
b = 1 + 4j
c = a + b  # c is now 3 + 7j
d = a * b  # d is now -5 + 11j

print(a,b,c,d)
```

```python
# You can use the real and imag attributes of a complex number to access its real and imaginary parts:
z = 3 + 4j
real_part = z.real  # real_part is now 3.0
imag_part = z.imag  # imag_part is now 4.0

print(z,real_part,imag_part)
```

```python
# You can use the abs function to get the magnitude of a complex number:
z = 3 + 4j
magnitude = abs(z)  # magnitude is now 5.0

print(z, magnitude)
```

```python
# You can use the cmath module to work with complex numbers in Python.
# This module provides functions for performing complex-valued math operations,
# such as trigonometric functions, exponential functions, and more:
import cmath

c = 3 + 4j

# Calculate the square root of c
sqrt_c = cmath.sqrt(c)

# Calculate the sine of c
sin_c = cmath.sin(c)

print(c,sqrt_c,sin_c)
```

```python
# To create a complex number, you can use the complex function,
# which takes two arguments: the real part and the imaginary part: 

c = complex(1, 2) # creates the complex number 1 + 2i

# Alternatively, you can use the j suffix to specify the imaginary part.
c = 1 + 2j # creates the complex number 1 + 2i

# Once you have a complex number, you can access its real and imaginary parts
# using the real and imag attributes, respectively.
real_part = c.real # gets the real part of c (1 in this case)
imaginary_part = c.imag # gets the imaginary part of c (2 in this case)

# You can also use the abs function to get the magnitude of a complex number,
# and the angle function to get its phase (in radians).
magnitude = abs(c) # gets the magnitude of c (sqrt(5) in this case)
phase = cmath.phase(c) # gets the phase of c (around 1.1071487177940904 in this case)

print(c, real_part, imaginary_part, magnitude, phase)
```

## Plots of Parabolas
***
A parabola is a graph of a quadratic function, which is a polynomial function of the form f(x) = ax^2 + bx + c, where a, b, and c are constants. The graph of a quadratic function is a curve that is shaped like a parabola. When a parabola is plotted on a graph, it has a definite vertex, which is the point where the curve changes direction. The axis of symmetry of a parabola is a line that divides the parabola into two mirror images. The vertex of a parabola is always on the axis of symmetry. When a parabola is plotted on a graph, it can open upwards or downwards, depending on the value of the constant a. If a is positive, the parabola opens upwards, and if a is negative, the parabola opens downwards. The curve of a parabola is always concave, which means that it is always curved inwards.

```python
# Import matplotlib and numpy.
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 8))

# Define the x values.
x = np.linspace(-10, 10, 100)

# Define the y values for the first parabola.
y1 = x**2

# Define the y values for the second parabola.
y2 = (x-1)**2

# Define the y values for the third parabola.
y3 = (x+2)**2

# Plot the parabolas.
plt.plot(x, y1, 'r--') # plot the first parabola in red with dashes
plt.plot(x, y2, 'g-.') # plot the second parabola in green with dots and dashes
plt.plot(x, y3, 'b:') # plot the third parabola in blue with dots
plt.title('Parabolas') # set the title
plt.xlabel('x') # set the x-axis label
plt.axhline(0.0, color='grey');
plt.ylabel('y') # set the y-axis label
plt.grid(True) # show a grid
plt.show()
```

## Quadratic Formula
***

The quadratic formula is a formula that allows you to solve for the roots of a quadratic equation, which is an equation of the form:

ax^2 + bx + c = 0

where a, b, and c are constants. The quadratic formula is:

x = (-b +/- sqrt(b^2 - 4ac)) / (2a)

To use the quadratic formula, you plug in the values of a, b, and c from your equation into the formula, and then solve for x.

For example, if you have the equation: 2x^2 + 5x - 3 = 0, then you can use the quadratic formula to solve for the roots (values of x that make the equation true). Plugging in the values, you get:

x = (-5 +/- sqrt(5^2 - 42(-3))) / (2*2)
= (-5 +/- sqrt(25 + 24)) / 4
= (-5 +/- sqrt(49)) / 4
= (-5 +/- 7) / 4
= (-12)/4 or (2)/4
= -3 or 0.5

So the roots of the equation are x = -3 and x = 0.5.

```python
#example of how you can use the quadratic formula in Python to solve a quadratic equation:


# Import the sqrt function from the math module.
from math import sqrt

# Define the constants a, b, and c.
a = 1
b = -3
c = 2

# Use the quadratic formula to find the values of x.
x1 = (-b + sqrt(b**2 - 4*a*c)) / (2*a)
x2 = (-b - sqrt(b**2 - 4*a*c)) / (2*a)

# Print the solutions.
print(x1)
print(x2)
```

## Fourier Transform
***

The Fourier transform is a mathematical technique that decomposes a function into its constituent frequencies. It is widely used in many areas of science and engineering, including signal processing, image analysis, and quantum mechanics.

The Fourier transform of a function f(t) is denoted by F(ω), where ω is the frequency. The Fourier transform can be thought of as a way to represent a function in terms of the frequencies that make it up.

One of the key properties of the Fourier transform is that it is reversible: the original function can be reconstructed from its Fourier transform. This makes it a powerful tool for analyzing and manipulating functions, as well as for solving differential equations.

There are several different types of Fourier transforms, including the continuous Fourier transform, the discrete Fourier transform, and the fast Fourier transform (FFT). The FFT is an efficient algorithm for computing the discrete Fourier transform, and is widely used in practice.

```python
# Heres an example of the Fast Fourier Transform (FFT) to analyze the frequency components of a signal.
# First, we generate a signal with two sinusoidal components and some noise:

from scipy.fft import fft

# Generate signal
fs = 1000  # sample rate
T = 1/fs  # sample period
t = np.arange(0, 1, T)  # time vector
f1 = 50  # frequency of first component
f2 = 120  # frequency of second component
y = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t) + 0.1*np.random.randn(t.size)

# Plot signal
plt.figure(figsize=(12, 8))
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Compute FFT
Y = fft(y)

# Get frequency vector
f = np.arange(0, fs, fs/t.size)

# Plot magnitude spectrum
plt.figure(figsize=(12,8))
plt.plot(f, np.abs(Y))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

## Inverse Fourier Transform
***
The inverse Fourier transform is the process of transforming a frequency domain representation of a signal back into its time domain representation. This can be useful for analyzing the time domain behavior of a signal that has been transformed into the frequency domain, or for synthesizing a signal from its frequency domain representation.

```python
# Define the frequency and sampling rate of the signal.
frequency = 10 # Hz
sampling_rate = 1000 # Hz

# Generate the time points for the signal.
time_points = np.linspace(0, 1, sampling_rate, endpoint=False)

# Generate the signal in the time domain.
signal = np.sin(2 * np.pi * frequency * time_points)

# Perform the Fourier transform of the signal.
fourier_transform = fft(signal)

# Perform the inverse Fourier transform to reconstruct the signal in the time domain.
reconstructed_signal = fft(fourier_transform)

# Plot the magnitude of the original and reconstructed signals.
plt.figure(figsize=(12,8))
plt.plot(time_points, np.abs(signal), label='Original')
plt.plot(time_points, np.abs(reconstructed_signal), label='Reconstructed')
plt.legend()
plt.show()
```

## Formula
***
In mathematics, the Fourier transform is a way of expressing a function as a sum of periodic components. It is used to analyze the frequency content of a signal or to filter a signal to remove or emphasize certain frequencies.

There are two forms of the Fourier transform: the continuous Fourier transform and the discrete Fourier transform.

__The continuous Fourier transform of a function f(t) is given by:__

F(ω) = ∫f(t)e^(-iωt)dt

where ω is the angular frequency, and i is the imaginary unit.

__The inverse continuous Fourier transform is given by:__

f(t) = ∫F(ω)e^(iωt)dω

The discrete Fourier transform (DFT) is a discrete version of the continuous Fourier transform. It is used to transform a finite-length sequence of samples into the frequency domain.

__The DFT of a sequence x[n] is given by:__

X[k] = ∑x[n]e^(-i2πkn/N)

where k is the frequency index, n is the time index, and N is the length of the sequence.

__The inverse DFT is given by:__

x[n] = (1/N) ∑X[k]e^(i2πkn/N)

NumPy provides functions for calculating the DFT and inverse DFT of a signal. The fft function calculates the DFT, and the ifft function calculates the inverse DFT.

__Here is an example of how to calculate the discrete Fourier transform of a signal using the DFT formula:__

```python
# Define the signal in the time domain.
signal = [1, 2, 3, 4, 5]

# Define the frequency indices.
k = np.arange(len(signal))

# Calculate the DFT of the signal.
fourier_transform = []
for n in range(len(signal)):
    sum = 0
    for k in range(len(signal)):
        sum += signal[k] * np.exp(-2j * np.pi * k * n / len(signal))
    fourier_transform.append(sum)

print(fourier_transform)  # [15. +0.j, -2.5 +3.4409548605386882j, -2.5 -1.449548605386881j, -2.5 +0.5504513946031182j, -2.5 -1.449548605386881j]
```

This code defines the signal in the time domain as an array of 5 samples, and calculates the DFT of the signal by summing over the time indices and the frequency indices. The resulting Fourier transform is a complex-valued array.

You can also use this formula to calculate the inverse DFT by replacing the exponent in the sum with its inverse and dividing the result by the length of the signal. Here is an example of how to calculate the inverse DFT:

```python
# Define the Fourier transform of the signal.
fourier_transform = [15, -2.5 + 3.4409548605386882j, -2.5 - 1.449548605386881j, -2.5 + 0.5504513946031182j, -2.5 - 1.449548605386881j]

# Calculate the inverse DFT to reconstruct the signal in the time domain.
reconstructed_signal = []
for n in range(len(fourier_transform)):
    sum = 0
    for k in range(len(fourier_transform)):
        sum += fourier_transform[k] * np.exp(2j * np.pi * k * n / len(fourier_transform))
    reconstructed_signal.append(sum)

print(reconstructed_signal)
```
